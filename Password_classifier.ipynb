```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas import Series, DataFrame
import seaborn as sns
%matplotlib inline
```

```python
# formatting in pandas dataframe to manipulate through pandas
passwords= pd.read_csv('/home/patel512/BANG!/Machine_Learn/DATA_SET/passwords.txt',header= None, skiprows=16, names=['password'])

# Creating a column called length which will show the length of password
passwords['length'] = passwords['password'].str.len() 
# eliminatin the data having password of length greater than
passwords= passwords[passwords.length<28]
```

```python
# filteration for password only containing digits

passwords['mask_char'] = passwords['password'].str.contains('^[0-9]+$') #0Creating a column mask_char  which only has numerical value

digit = (passwords[passwords['mask_char'] == True]).drop('mask_char', axis = 1)

digi_count=len(digit)    # count of password which only has character
print(digi_count)
```

```python
# filtration for passwords only containing character

passwords['mask_digit'] = passwords['password'].str.contains('^[a-zA-Z]+$') # creating new column which will give true value only if the password is character only 

char = (passwords[passwords['mask_digit'] == True]).drop('mask_digit', axis = 1)

char_count=len(char)  # count of pass which only has character
print(char_count)
```

```python
# filtration for password containing special symbol

passwords['special_symbol'] = passwords['password'].str.contains('^[-!$%^&*()_+|~=`{}\[\]:";\'\<>?,.\/]+$')

symbol = (passwords[passwords['special_symbol'] == True]).drop('special_symbol', axis = 1)
# symbol for displaying password containing special symbol

symbol_count = len(symbol)
print(symbol_count)
```
